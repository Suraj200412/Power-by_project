```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Rainfall_Data_Germany_Complete.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
min_temp = df["Temperature (°C)"].min()
print(f"Minimum Temperature: {min_temp} °C")
```

```python
max_temp = df["Temperature (°C)"].max()
print(f"Maximum Temperature: {max_temp} °C")
```

```python
avg_temp = df["Temperature (°C)"].mean()
print(f"Average Temperature: {avg_temp:.2f} °C")
```

```python
temps = {"Min Temp": min_temp, "Max Temp": max_temp, "Avg Temp": avg_temp}
plt.figure(figsize=(6,5))
plt.bar(temps.keys(), temps.values(), color=['lightblue', 'salmon', 'lightgreen'])
plt.title("Temperature Summary")
plt.ylabel("Temperature (°C)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,8))
df.groupby("City")["Rainfall (mm)"].mean().sort_values(ascending=False).plot(kind="bar", color='red')
plt.title("Average Rainfall by City (mm)")
plt.xlabel("City")
plt.ylabel("Average Rainfall (mm)")
plt.xticks(rotation=40)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
df.groupby("Year")["Rainfall (mm)"].mean().plot(marker='o', color='green')
plt.title("Average Rainfall Over the Years")
plt.xlabel("Year")
plt.ylabel("Average Rainfall (mm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
df.groupby("Month")["Rainfall (mm)"].mean().plot(kind="bar", color='cornflowerblue')
plt.title("Average Rainfall by Month", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Average Rainfall (mm)")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
