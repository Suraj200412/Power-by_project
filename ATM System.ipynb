```python
import os
import mysql.connector

conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='',
    database='ATM'
)
cursor = conn.cursor()

def file(username):
    return f"{username}.txt"

def register():
    name = input("Enter your name: ").strip()
    pin = input("Enter your pin: ").strip()

    try:
        query = "INSERT INTO customer(pin, name, balance) VALUES (%s, %s, %s)"
        values = (pin, name, 0)
        cursor.execute(query, values)
        conn.commit()

        if os.path.exists(file(name)):
            print("This user is already registered (file exists).")
            return

        with open(file(name), "w") as f:
            f.write(f"name: {name}\n")
            f.write(f"pin: {pin}\n")
            f.write("balance: 0\n")

        print("Registration successful!")

    except mysql.connector.IntegrityError as e:
        if "Duplicate entry" in str(e):
            print("This user is already registered in the database.")
        else:
            print("Database error:", e)

def login():
    name = input("Enter your name: ").strip()
    pin = input("Enter your pin: ").strip()

    if not os.path.exists(file(name)):
        print("User not found. Please register first.")
        return

    with open(file(name), "r") as f:
        data = f.readlines()
        saved_name = data[0].split(":")[1].strip()
        saved_pin = data[1].split(":")[1].strip()
        balance = int(data[2].split(":")[1].strip())

    if name == saved_name and pin == saved_pin:
        print("Login successful!")

        while True:
            print(f"\nYour balance: ₹{balance}")
            print("""
    1. Check Balance
    2. Deposit Money
    3. Withdraw Money
    4. Exit
            """)
            choice = input("Enter your choice (1–4): ").strip()

            if choice == "1":
                print(f"Your balance: ₹{balance}")

            elif choice == "2":
                amount = int(input("Enter amount to deposit: "))
                if amount <= 0:
                    print("Invalid amount!")
                else:
                    balance += amount
                    print(f"₹{amount} credited to your account.")

            elif choice == "3":
                amount = int(input("Enter amount to withdraw: "))
                if amount <= 0:
                    print("Invalid amount!")
                elif amount > balance:
                    print("Insufficient balance!")
                else:
                    balance -= amount
                    print(f"₹{amount} debited from your account.")

            elif choice == "4":
                with open(file(name), "w") as f:
                    f.write(f"name: {name}\n")
                    f.write(f"pin: {pin}\n")
                    f.write(f"balance: {balance}\n")

                print("Thank you for using the ATM. Goodbye!")
                break

            else:
                print("Enter a valid input (1–4).")

    else:
        print("Name or PIN do not match.")

def menu():
    while True:
        print("\n1. Register\n2. Login\n3. Exit")
        c = input("Enter your choice: ").strip()

        if c == "1":
            register()
        elif c == "2":
            login()
        elif c == "3":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

menu()
```
